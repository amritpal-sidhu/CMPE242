```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
adc_data = np.genfromtxt("adc_log_data_original.csv")
time_data_ms = np.arange(0, 1.024, 1e-3)
```

```python
print(adc_data.shape)
print(time_data_ms.shape)
```

```python
plt.scatter(time_data_ms, adc_data)
plt.title("ADC data w.r.t. time")
plt.xlabel("Time (s)")
plt.ylabel("ADC value")
plt.show()
```

```python
adc_fft = np.fft.fft(adc_data)
adc_power_spectrum = np.abs(adc_fft)**2
freq = np.fft.fftfreq(adc_data.size, d=1e-3)
```

```python
plt.plot(freq, adc_fft.real, freq, adc_fft.imag)
plt.title("FFT of ADC data")
plt.xlabel("Frequency")
plt.ylabel("Amplitude")
plt.show()
```

```python
plt.plot(freq, adc_power_spectrum)
plt.title("Power Spectrum of ADC data")
plt.xlabel("Frequency")
plt.ylabel("Power")
plt.show()
```
